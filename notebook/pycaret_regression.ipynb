# Pandas Profilling

```python
# Load the dataset
from pycaret.datasets import get_data
house_prices = get_data('house', profile=True)
```

# Setup

```python
from pycaret.regression import *
# Initialize the setup
reg1 = setup(data=house_prices, target='SalePrice', session_id=123, experiment_name='Regression')
```

# Compare Model

```python
model_terbaik = compare_models(sort='MSE')
```

```python
print(model_terbaik)
```

# Create Model

Note : 

Ini bertujuan untuk membuat model yang akan saya gunakan dalam melakukan training pada dataset saya

# Tune Model

```python
cb = create_model('catboost')
tune_cb = tune_model(cb, n_iter=50, optimize='MSE', search_library='optuna', search_algorithm='random')
```

# Evaluate Model

```python
plot_model(tune_cb)
```

```python
plot_model(tune_cb, plot='feature')
```

# Interpret Model

```python
interpret_model(tune_cb)
```

# Save Model

```python
save_model(tune_cb, 'best_model_regression')
```

# Load Best Model

```python
load_best = load_model('best_model_regression')
print(load_best)
```
